```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:\\Users\\luiss\\OneDrive\\Quant\\Financial Data\\EURUSD_Fsb\\EURUSD_H1.csv", parse_dates=["timestamp"])
df = df.drop("volume", axis=1)
df.set_index("timestamp", inplace=True)
df.head()
```

```python
def find_swings(df, window=3):
    highs, lows = [], []
    for i in range(window, len(df) - window):
        local_high = df["high"].iloc[i] == max(df["high"].iloc[i-window:i+window+1])
        local_low = df["low"].iloc[i] == min(df["low"].iloc[i-window:i+window+1])
        if local_high:
            highs.append((df.index[i], df["high"].iloc[i]))
        if local_low:
            lows.append((df.index[i], df["low"].iloc[i]))
    return highs, lows

highs, lows = find_swings(df, window=3)
```
